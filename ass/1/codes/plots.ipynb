```python
import matplotlib.pyplot as plt
import numpy as np

s_logs = './logs/s_logs'
d_logs = './logs/d_logs'
f1 =  open(s_logs, 'r')
f2 =  open(d_logs, 'r')
line_s = f1.readlines()
line_d = f2.readlines()
f2.close()
f2.close()

s_vm = {'mat_v1':[], 'mat_v2':[], 'mat_cache_v2':[], 'mat_v3':[], 'mat_cache_v3':[], 'mat_v4':[], 'mat_v5':[], 'mat_v6':[]}
key = ''
for l in line_s:
    if l.startswith('f'):
        key = l.split(' ')[1][:-1]
    if l.startswith('D'):
        t = float(l.split(' ')[2])
        s_vm[key].append(t)
        
d_vm = {'mat_v1':[], 'mat_v2':[], 'mat_cache_v2':[], 'mat_v3':[], 'mat_cache_v3':[], 'mat_v4':[], 'mat_v5':[], 'mat_v6':[]}
key = ''
for l in line_d:
    if l.startswith('f'):
        key = l.split(' ')[1][:-1]
    if l.startswith('D'):
        t = float(l.split(' ')[2])
        d_vm[key].append(t)
```

```python
# Single-VM
processes = [1, 2, 4, 8, 16, 32]
plt.figure()
plt.title('Single-VM')
plt.xticks(processes)
plt.yticks(np.linspace(0, 0.5, 12))
plt.plot([2, 4, 8, 16, 32], s_vm['mat_v1'], label='mat_v1')
for k in s_vm:
    if k == 'mat_v1':
        continue
    plt.plot(processes, s_vm[k], label=k)
plt.grid()
plt.legend(loc='best', ncol=2)
plt.show()
```

```python
# Double-VM
processes = [1, 2, 4, 8, 16, 32]
plt.figure()
plt.title('Double-VM')
plt.xticks(processes)
plt.yticks(np.linspace(0, 31, 12))
plt.plot([2, 4, 8, 16, 32], d_vm['mat_v1'], label='mat_v1')
for k in d_vm:
    if k == 'mat_v1':
        continue
    plt.plot(processes, d_vm[k], label=k)
plt.grid()
plt.legend(loc='best', ncol=2)
plt.show()
```

```python
# Double-VM_1-4
processes = [1, 2, 4]
plt.figure()
plt.title('Double-VM_1-4')
plt.xticks(processes)
plt.yticks(np.linspace(0, 0.5, 12))
plt.plot([2, 4], d_vm['mat_v1'][:2], label='mat_v1')
for k in d_vm:
    if k == 'mat_v1':
        continue
    plt.plot(processes, d_vm[k][:3], label=k)
plt.grid()
plt.legend(loc='best', ncol=2)
plt.show()
```

```python
# Double-VM_8-32
processes = [8, 16, 32]
plt.figure()
plt.title('Double-VM_8-32')
plt.xticks(processes)
plt.yticks(np.linspace(0, 31, 12))
plt.plot([8, 16, 32], d_vm['mat_v1'][2:], label='mat_v1')
for k in d_vm:
    if k == 'mat_v1':
        continue
    plt.plot(processes, d_vm[k][3:], label=k)
plt.grid()
plt.legend(loc='best', ncol=2)
plt.show()
```

```python
# Double-VM v2, 2-cache, 3
processes = [1, 2, 4, 8, 16, 32]
plt.figure(figsize=(16, 4))
plt.title('Double-VM v2, 2-cache, 3')
plt.xticks(processes)
plt.yticks(np.linspace(0, 10, 12))
for k in d_vm:
    if k in ['mat_v1', 'mat_cache_v3', 'mat_v4', 'mat_v5', 'mat_v6']:
        continue
    plt.plot(processes, d_vm[k], label=k)
plt.grid()
plt.legend(loc='best', ncol=2)
plt.show()
```
